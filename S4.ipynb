```python
# Phan loai nhom tuoi cua data trong info_df.
# Tao mot cot moi trong info_df ten la Age_Group. Rules:
# Age < 18 tuoi : 'Kid'
# 18 <= Age < 30: 'Teenager'
# 30 <= Age < 55: 'Middle_Age'
# Age >= 55: 'Old'
age_list = []
for i in range(len(info_df)):
    if info_df['Age'][i] < 18:
        age_list.append('Kid')
    elif info_df['Age'][i] < 30: # >=18
        age_list.append('Teenager')
    elif info_df['Age'][i] < 55: # >=30
        age_list.append('Middle_Age')
    else: # >=55 or NaN
        age_list.append('Old')
info_df['Age_Group'] = age_list
info_df
```
