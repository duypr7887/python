```python
# Group "Age" column into 4 groups, create a new column "Age_Group":
# Age < 18: "kid"
# 18 <= Age < 30: "teenager"
# 30 <= Age < 55: "middle_age"
# Age >= 55: "elderly"
age_group = []
for age in df['Age']:
    if age < 18:
        age_group.append('kid')
    elif age < 30:
        age_group.append('teenager')
    elif age < 55:
        age_group.append('middle_age')
    else:
        age_group.append('elderly')
        
df['Age_Group'] = age_group
```

```python
# Write a function named check_palindrome that takes in a positive integer as input
# This function should return True if the inputted number is a Palindrome number
# For example: check_palindrome(12321) should return True
#              check_palindrome(1242) should return False
def check_palindrome(n):
    num = n
    reversed_num = ''
    while num > 0:
        last_num = num % 10
        reversed_num = reversed_num + str(last_num)
        num = num // 10
    if reversed_num == str(n):
        return True
    else:
        return False
```
