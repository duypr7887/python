```python
def cal_mean(l):
    s = 0
    for i in l.dropna():
        s = s + i
    avg = s/len(l)
    return round(avg,2)
```

```python
cal_mean(df_age_cabin['Age'])
```

```python
df_age_cabin['Age'].fillna(value = cal_mean(df_age_cabin['Age']), inplace = True)
```
