```python
# Buoi so 1
# - Data type: dạng giá trị của dữ liệu (integer, float, string, datetime, boolean...)
# - Data structure: cách dữ liệu được lưu trữ (tabular, non-tabular)
# - Operators: 
# + Arithmetic operators: +, -, *, /, **, //, %
# + Comparison operators: >, <, >=, <=, ==, !=
# + Logical operators: and, or, in, not in
# - Casting: chuyển đổi 1 dạng dữ liệu sang 1 dạng dữ liệu khác (convert)
# - Casting functions: str(); int(); float(); list();...
# - Built-in functions: print(); sum(); max(); min(); range(); type(); ...
# - String indexing: đi từ số bên trái đến số bên phải trừ đi 1 đơn vị. Ví dụ:
# s = 'abc123' => s[0:2] = 'ab'
```

```python
# Buoi so 2
# - If statement:
# if <condition>:
#     <do smth>
# elif <condition>:
#     <do smth>
# else:
#     <do smth>
# - For loop:
# for <element> in <iterable>:
#     <do smth>
# iterable: list, array, range, pandas column, etc.
# - While loop:
# while <condition>:
#     <do smth>
# - User-defined function: các hàm số mà người viết code tự tạo ra
def find_max(a,b):
    if a > b:
        return a
    else:
        return b

find_max(3,5)
# - Method and function:
# + Syntax:
#     function: <function>(<input>)
#     method: <input>.<method>(<params>)
# + Purpose: method tác động lên input; function trả ra output tương ứng với input
```

```python
# From a list of names, return a list of initials
def get_initials(s):
    initial_list = []
    for e in s.split(' '):
        initial_list.append(e[0])
    output = ''.join(initial_list)
    return output

get_initials('Nguyen Thanh Dat') # should return 'NTD'
```

```python
# Read name.xlsx into a dataframe, then create a new column named "Initials" which contains
# the initials of names in the "Name" columns.
df = pd.read_excel('name.xlsx', sheet_name = 'Sheet1')
df
```

```python
# Cach 1
l = []
for i in range(len(df)):
    l.append(get_initials(df['Name'][i]))

df['Initials'] = l
```

```python
df
```

```python
# Cach 2
df['Initials'] = df['Name'].apply(get_initials)
```
